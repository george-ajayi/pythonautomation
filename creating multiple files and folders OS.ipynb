```python
#CREATING A FOLDER AND SUB FOLDER 
desktop=os.path.expanduser('~/Desktop/').replace("\\","/")
df=pd.read_excel('~/Downloads/ds_salaries.xlsx')
# empt_types=list(df.Country.unique())
df2=df.query("job_title==['Data Scientist','Data Engineer','Data Analyst']")
job_titles=df2.job_title.unique()
experience_level=list(df2.experience_level.unique())
for title in job_titles:
    # CREATING FOLDER
    df3=os.mkdir(desktop+'Employment_type'+'/'+f'{title}')
    # CREATING SUB FOLDER
    df4=os.mkdir(desktop+'Employment_type'+'/'+f'{title}'+'/'+'Experience level')
    # LOADING FILES TO EACH SUB FOLDER
    da=df2.query(f"job_title=='{title}'")
    for exp in experience_level:
        data2=da.query(f"experience_level=='{exp}'")
        data2.to_excel(f'~/Desktop/Employment_type/{title}/Experience level'+'/'+f'{exp}.xlsx',index=False)



```
